```python
import pandas as pd
```

```python
df = pd.read_csv("nba_games.csv", index_col=0)
df
```

```python
def convert_to_seconds(mp):
    try:
        minutes, seconds = map(int, mp.split(':'))
        return minutes * 60 + seconds
    except ValueError:
        return None
```

```python
df['MP'] = df['MP'].apply(convert_to_seconds)
```

```python
df = df.drop("player", axis=1)
df = df.sort_values(by=['Player Name','date'])
df = df.drop_duplicates(subset=['Player Name', 'date'])
df = df.reset_index(drop=True)
df = df.fillna(0)
df
```

```python
def add_last_5(player_df, column, n=6):
    for i in range(1, n):
        shifted_values = player_df[column].shift(i)
        shifted_values = shifted_values.fillna(player_df[column])
        player_df[f"{column}_{i}_ago"] = shifted_values
    return player_df
```

```python
def add_last_5_av(player_df, column, n=7):
    for i in range(3, n):
        sum_list = player_df[[f"{column}_{j}_ago" for j in range(1, i)]].sum(axis=1)
        player_df[f"av_{column}_{i-1}_ago"] = round(sum_list / (i-1))
    return player_df
```

```python
non_stat_columns = ['date','season','team','Player Name']
stat_columns = df.columns[~df.columns.isin(non_stat_columns)]
stat_columns
```

```python
for column in stat_columns:
    df = df.groupby("Player Name", group_keys=False).apply(add_last_5,column=column)
    df = df.groupby("Player Name", group_keys=False).apply(add_last_5_av,column=column)
```

```python
df = df.copy()
```

```python
def add_target(player):
    player["target"] = player["PTS"].shift(-1,fill_value=round(player['PTS'].mean()))
    return player
```

```python
df = df.groupby("Player Name", group_keys=False).apply(add_target)
```

```python
df['next_home'] = df['home'].shift(-1,fill_value=2)
```
